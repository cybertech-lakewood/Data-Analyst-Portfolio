IF STATEMENT

```python
#a conditional statement used to decide wheter a certain statement will be executed or no
#if <condition>:
#    <code>
#elif <condition>:
#     <code>
#....
#else:
#    <code>
```

```python
age = 15

if age >= 23:
  print("You are old enough to party and drink")
elif age >= 14:
  print("You are old enogh to party but not drink")
else:
  print("You are not allowed to party")
```
